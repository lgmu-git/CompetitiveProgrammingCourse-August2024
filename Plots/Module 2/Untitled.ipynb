```python
import matplotlib.pyplot as plt
```

```python
# Define coefficients of the polynomial
a = 1
b = 10
c1 = 500
c2 = 1000
c3 = 1500
```

```python
# Generate positive x values
x = range(0, 5)  # Range of x values from 0 to 10
file = "poly01.png"
```

```python
# Generate positive x values
x = range(0, 50)  # Range of x values from 0 to 10
file = "poly02.png"
```

```python
# Calculate corresponding y values
y1 = [a*x_val**2 + b*x_val + c1 for x_val in x]
y2 = [a*x_val**2 + b*x_val + c2 for x_val in x]
y3 = [a*x_val**2 + b*x_val + c3 for x_val in x]

# Plotting
plt.figure(figsize = (8, 6))  # Optional: adjust figure size
plt.plot(x, y1, label=f'$y = {a}x^2 + {b}x + {c1}$', color='#009729', linewidth=2)  # Adjust color, label, linewidth as needed
plt.plot(x, y2, label=f'$y = {a}x^2 + {b}x + {c2}$', color='#E0DD16', linewidth=2)  # Adjust color, label, linewidth as needed
plt.plot(x, y3, label=f'$y = {a}x^2 + {b}x + {c3}$', color='#DD3E00', linewidth=2)  # Adjust color, label, linewidth as needed
plt.xlabel('x')  # Label for x-axis
plt.ylabel('f (x)')  # Label for y-axis
#plt.title("")  # Title of the plot
plt.grid(True)  # Optional: add grid
#plt.legend()  # Optional: add legend
# # 'upper left', 'upper right', 'lower left', 'lower right', 'upper center', 'lower center', 'center left', 'center right', 'center', 'best'
plt.legend(loc='lower left', bbox_to_anchor=(0, 1))  # Place legend
plt.axhline(0, color='black', linewidth=0.5)  # Optional: add horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5)  # Optional: add vertical line at x=0

# Save the plot in high quality
plt.savefig(file, dpi=300, bbox_inches='tight')  # Adjust file name and dpi (dots per inch) as needed
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Generate positive x values
x = [x * .1 for x in range(0, 200)]
file = "bigo02.png"


# Calculate corresponding y values
y0 = [1 if (x_val >= 0) else None for x_val in x]
y1 = [x_val**1 if (x_val >= 0) else None for x_val in x]
y2 = [x_val**2 if (x_val >= 0) else None for x_val in x]
y3 = [x_val**3 if (x_val >= 0) else None for x_val in x]
y4 = [math.log(x_val) if (x_val > 0) else None for x_val in x]


# Plotting
plt.figure(figsize = (16, 9))  # figure size
plt.plot(x, y0, label=f'$y = 1$',       color='#0000ff', linewidth=2)  # 
plt.plot(x, y1, label=f'$y = x$',       color='#00E5E5', linewidth=2)  # 
plt.plot(x, y2, label=f'$y = x^2$',     color='#009729', linewidth=2)  # 
plt.plot(x, y3, label=f'$y = x^3$',     color='#E0DD16', linewidth=2)  # 
plt.plot(x, y4, label=f'$y = \log(x)$', color='#DD3E00', linewidth=2)  # 
plt.xlabel('x')  # Label for x-axis
plt.ylabel('f (x)')  # Label for y-axis

plt.axis([0.5, 20, -0.5, 15])
plt.xticks(range(1, 20, 1))

#plt.title("")  # Title of the plot
plt.grid(True)  # Optional: add grid
# # 'upper left', 'upper right', 'lower left', 'lower right', 'upper center', 'lower center', 'center left', 'center right', 'center', 'best'
#plt.scatter(1, 12, color='blue', label='c')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend

plt.axhline(0, color='black', linewidth=0.5)  # Optional: add horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5)  # Optional: add vertical line at x=0

# Save the plot in high quality
plt.savefig(file, dpi=300, bbox_inches='tight')  # Adjust file name and dpi (dots per inch) as needed
plt.show()
```
